```python

# read data from CSV file
df = pd.read_csv("AMS_metadata.csv", header=None, engine='python')

# extract data columns
x = np.array(df.iloc[:, 0])
y = np.array(df.iloc[:, 0])

# plot the data
plt.scatter(x, y)
plt.xlabel('DAYS')
plt.ylabel('RAIN(mm)')
plt.title('Rain in Amsterdam')
plt.show()
```

```python


# calculate coefficients for line of best fit
coefficients = np.polyfit(x, y, 1)
m = coefficients[0]
b = coefficients[1]

# create scatter plot and line of best fit
plt.scatter(x, y)
plt.plot(x, m*np.array(x) + b, color='red')

# Show the plot
plt.show()
```
